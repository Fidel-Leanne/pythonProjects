```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt
```

```python
#load data

df= pd.read_csv('C:/Users/Operator/Desktop/CODING/Python/.ipynb_checkpoints/DataHouse/New folder/mines.csv')
df
```

```python
# Load data

# Use only the specified variables and drop the target
X_clustering = df[['Var1', 'Var3', 'Var5']]
X_clustering

```

```python
df.describe()
```

```python
# Standardize the data (often necessary for clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_clustering)
```

```python
# Using KMeans with, for instance, 3 clusters
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters)
plt.title('Clusters Visualization')
plt.show()
```

Multiple linear Regression

```python
# Assuming data is already loaded from the previous section
X = df.drop('Target', axis=1)
y = df['Target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

# Calculate and print the mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```
