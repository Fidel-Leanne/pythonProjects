Naive Bayes 

Industrial use

-Text Classification: One of the classic and most popular applications of Naïve Bayes is in text classification, particularly in:

Spam Filtering: Differentiating between spam and non-spam emails.
Sentiment Analysis: Classifying sentiments of textual data as positive, negative, or neutral.
Document Categorization: Assigning categories or topics to text documents.
Real-time Prediction: Naïve Bayes is a relatively fast algorithm, making it suitable for real-time predictions, for example, categorizing news articles as they are published.

-Recommendation Systems: While more complex algorithms like collaborative filtering or deep learning models are often preferred for recommendation systems, Naïve Bayes can be employed in simpler recommendation systems, especially when data is sparse.

-Medical Diagnosis: It can be used to diagnose diseases based on a combination of symptoms. Given symptoms, it can calculate the probability of various diseases.

-Weather Prediction: Predicting whether it will rain, be sunny, etc., based on various parameters.

-Face Recognition: While more complex models have surpassed it, Naïve Bayes has been used in face recognition, especially in its early stages.

-Credit Scoring: Predicting whether a customer is a high or low credit risk based on various attributes.

-Customer Classification: Classifying customers into various segments based on their buying behavior.

-Fault Detection: Detecting faults in manufacturing processes or machinery by analyzing various symptoms or signals.

-Trending Topics: In social media platforms, Naïve Bayes can be used to identify and categorize trending topics.

```python
df= pd.read_csv('C:/Users/Operator/Desktop/CODING/Python/.ipynb_checkpoints/DataHouse/diabetesDataset/diabetes.csv')
```

```python
df= df.astype(float)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
df.columns
```

```python
X= df.drop('Outcome', axis=1)
y=df['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
# Initialize GaussianNB
clf = GaussianNB()

# Train the classifier
clf.fit(X_train, y_train)

# Predict on the test data
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
