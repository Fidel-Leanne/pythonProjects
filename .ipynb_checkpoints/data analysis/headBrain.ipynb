```python
import sys 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
import scipy 
import sklearn as sk
```

```python
plt.pyplot.rcParams['figure.figsize']=(20,10)

data=pd.read_csv('C:/Users/Operator/Desktop/CODING/Python/.ipynb_checkpoints/DataHouse/headbrain/headbrain.csv')
data.shape
data
```

```python
#Collecting X and Y
X= data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)

#TOTAL NUMBER OF VALUES
n = len(X)


#using the forrmula to calculate b1 and b0

numer=0
denom=0

for i in range(n):
    numer+=(X[i]-mean_x)*(Y[i] - mean_y) #finding numerator of equation 2
    denom+=(X[i]-mean_x)**2

b1=numer/denom
b0=mean_y-(b1*mean_x)

print(b1, b0)
```

```python
sns.scatterplot(data=data)
```

```python
import matplotlib.pyplot as plt
```

```python
max_x= np.max(X)+100
min_x = np.min(X)-100

x = np.linspace(min_x, max_x, 1000)
y = b0 + b1 * x

plt.plot(x, y, color="blue", label='Regression line')  # Use plt.plot() directly
plt.scatter(X, Y, c='red', label="scatter plot")
plt.legend()
plt.show()  # To display the plot

```

```python
ss_t= 0
ss_r=0


for i in range(n):
 y_pred=b0+b1*X[i]
 ss_t+=(Y[i]-mean_y)**2
 ss_r+=(Y[i]-y_pred)**2

r2=1-(ss_r/ss_t)
print(r2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X=X.reshape((n,1))

reg=LinearRegression()

reg=reg.fit(X,Y)

Y_pred=reg.predict(X)

r2_score= reg.score(X,Y)
print("R-Squared: ", r2_score )
```
