```python
# Import necessary libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict, KFold, StratifiedKFold


var1= df['Var1'].values
Target= df['Target'].values
# Data
X = np.array(var1)
Y = np.array(Target)
X= X.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)



# # Create a linear regression model
model = LinearRegression()

```

```python
# # Fit the model to the data
model.fit(X, Y)

# # Predict the Y values
Y_pred = model.predict(X)

# # Print the slope (m) and y-intercept (c)
print("Slope (m):", model.coef_[0])
print("Y-intercept (c):", model.intercept_)

# # Visualize the data points and the regression line
plt.scatter(X, Y, color='blue', label="Actual Data Points")
plt.plot(X, Y_pred, color='red', label="Regression Line")
plt.xlabel("Target")
plt.ylabel("Var 1")
plt.title("target vs Var 1")
plt.legend()
plt.grid(True)
plt.show()
```

```python
b0=0
b1=0
n = len(X)

mean_x=np.mean(X)
mean_y=np.mean(Y)

numer=0
denom=0

for i in range(n):
    numer+=(X[i]-mean_x)*(Y[i] - mean_y) #finding numerator of equation 2
    denom+=(X[i]-mean_x)**2

b1=numer/denom
b0=mean_y-(b1*mean_x)

print( "b1:",b1,"b0:", b0)

ss_t= 0
ss_r=0


for i in range(n):
 y_pred=b0+b1*X[i]
 ss_t+=(Y[i]-mean_y)**2
 ss_r+=(Y[i]-y_pred)**2

r2=1-(ss_r/ss_t)
print(r2)
```
